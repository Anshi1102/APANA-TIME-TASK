```python
import pandas as pd

# Load the dataset
df = pd.read_excel("Data set (2).xlsx")

# Show initial nulls
print("Null values before handling:\n", df.isnull().sum())

# Impute missing values using MEAN (recommended for continuous numeric features)
for col in df.select_dtypes(include='number').columns:
    df.fillna({col: df[col].mean()}, inplace=True)

# Show nulls after handling
print("\nNull values after handling:\n", df.isnull().sum())

# --- Exploratory Data Analysis (EDA) ---

# Central Tendency
print("\n--- Central Tendency ---")
print("Mean:\n", df.mean(numeric_only=True))
print("Median:\n", df.median(numeric_only=True))
print("Mode:\n", df.mode(numeric_only=True).iloc[0])

# Measures of Dispersion
print("\n--- Dispersion ---")
print("Variance:\n", df.var(numeric_only=True))
print("Standard Deviation:\n", df.std(numeric_only=True))
print("Range:\n", df.max(numeric_only=True) - df.min(numeric_only=True))

# Shape of Distribution
print("\n--- Skewness ---")
print(df.skew(numeric_only=True))

print("\n--- Kurtosis ---")
print(df.kurt(numeric_only=True))

# Optional: Save cleaned data
df.to_excel("Cleaned_Data.xlsx", index=False)
```
